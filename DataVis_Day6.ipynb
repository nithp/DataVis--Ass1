EXPLORING PLOTLY EXPRESS 

```python
import plotly.express as px
```

```python
df=px.data.iris()
fig=px.scatter(df, x='sepal_width', y='sepal_length',color='species')
fig.show()
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length")
fig.show()
```

```python
fig=px.scatter(df, x='sepal_width', y='sepal_length',color='species',marginal_y ='violin', marginal_x='box',trendline='ols', template="simple_white")
fig.show()
```

```python
#Normalisation
df['e'] = df['sepal_width']/100
fig=px.scatter(df, x='sepal_width', y='sepal_length',color='species',error_y ='e', error_x='e')
fig.show()
```

```python
d= px.data.tips()
fig=px.bar(d, x='sex', y='total_bill',color='smoker',barmode="group")
fig.show()
```

```python
fig = px.pie(d, values='tip', names='day')
fig.show()
```

```python
fig = px.box(d, x="time", y="total_bill", points="all")
fig.show()
```

```python
fig=px.box(d, x='day', y='total_bill',color='smoker',notched= True)
fig.show()
```

```python
fig = px.violin(d, y="total_bill",points='all')
fig.show()
```

```python
fig=px.violin(d, x='smoker', y='tip',color='sex',box= True, points='all')
fig.show()
```

```python
fig=px.strip(d, x='total_bill', y='time',color='smoker', orientation='h')
fig.show()
```

```python
fig=px.density_contour(d, x='total_bill', y='time',color='smoker', marginal_y='histogram')
fig.show()
```

```python
fig = px.histogram(d, x="total_bill")
fig.show()
```

```python
fig = px.histogram(d, x="total_bill", nbins=20)
fig.show()
```

```python
fig = px.histogram(d, x="total_bill", color="sex")
fig.show()
```
